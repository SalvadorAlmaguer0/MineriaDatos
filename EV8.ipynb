```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Cargamos los datos
data = pd.read_csv('real_estate.csv')
# definimos las variables
y = data['price']
x1 = data['size']
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 223.1787*x1 + 101900
fig = plt.plot(x1, yhat,lw=4, c='orange', label = 'regression line')
plt.xlabel('Tama√±o', fontsize = 20)
plt.ylabel('Precio', fontsize = 20)
plt.show()
```

```python
# Predicci√≥n de una propiedad que mide 900
tam = 900 # Tama√±o
precio = 223.1787*(tam) + 101900


plt.scatter(x1,y)
yhat = 223.1787*x1 + 101900
fig = plt.plot(x1, yhat,lw=3, c='orange')
plt.plot(tam, precio,marker="o", c='red', label = round(precio,2))
plt.xlabel('Tama√±o', fontsize = 20)
plt.ylabel('Precio', fontsize = 20)
plt.legend()
plt.show()

# En rojo se muestra el valor de nuestra preduccion (302,276.83)
```
