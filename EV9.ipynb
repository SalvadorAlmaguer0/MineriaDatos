```python
# Segmentacion del mercado
# Ejemplo de agrupamiento
# Importar las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
# Importat el modulo de KMeans
from sklearn.cluster import KMeans
```

```python
# Cargar datos
data = pd.read_csv('Encuesta.csv')
data
```

```python
# Graficar los datos
plt.scatter(data['Satisfaccion'], data['Lealtad'])
# Se nombran los ejes
plt.xlabel('Satisfaccion')
plt.ylabel('Lealtad')
```

```python
# Creamos una copia de data y la guardamos en x
x = data.copy()
```

```python
# Agrupamiento
# Creamos un objeto al que llamamos kmeans
# El numero en los parametros es K, o el numero de clusters

kmeans = KMeans(2)
# Ajustamos los datos en .fit
kmeans.fit(x)
```

```python
# Resultados de agrupamiento
# Creamos una copia de los datos de entrada.
grupos = x.copy()
# Take note of the predicted clusters
grupos['grupos_pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(grupos['Satisfaccion'], grupos['Lealtad'], c=grupos['grupos_pred'], cmap = 'rainbow')
plt.xlabel('Satisfaccion')
plt.ylabel('Lealtad')
```

```python
# Estandarizar las variables
# Se va a estandarizar las variables y veremos los resultados.
# Important la libreria
from sklearn import preprocessing
# Escalar las entradas 
# preprocessing.scale escala cada variable (columna en x) con relacion a si misma
# The new result is on array
x_escalada = preprocessing.scale(x)
x_escalada
```

```python
# Realizar el metodo del codo
wcss = []

# Crear todas las posibles soluciones de agrupamiento con un ciclo.
# Se eligio una solucion de 1 a 9 grupos, esto puede variar.
for i in range(1,10):
    # NUmero de  clusters en variable (i)
    kmeans = KMeans(i)
    # Ajustar los datos ESTANDARIZADOS
    kmeans.fit(x_escalada)
    # Abrir el wcss variable para la iteracion
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
# Graficar el numero de grupos o clusters vs WCSS
plt.plot(range(1,10),wcss)
# Nombrar los ejes
plt.xlabel('Numero de grupos o clusters')
plt.ylabel('WCSS')
```

```python
# Numero de grupos
kmeans_nuevos = KMeans(4)

kmeans_nuevos.fit(x_escalada)
# Crear otra hoja de datos con los grupos o clusters predicticos.
grupos_nuevos = x.copy()
grupos_nuevos['grupo_pred'] = kmeans_nuevos.fit_predict(x_escalada)
```
