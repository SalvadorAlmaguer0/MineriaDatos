```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data2 = pd.read_csv('AlumnosAdmision.csv')
data2
```

```python
data2['Attendance'] = data2['Attendance'].map({'Yes': 1 ,'No' : 0})
data2
```

```python
y=data2['GPA']
x1=data2[['SAT', 'Attendance']]
x = sm.add_constant(x1)
```

```python
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data2['SAT'],y)

yhat_no = 0.6439 + 0.0014 * data2['SAT']
yhat_yes = 0.8665 + 0.0014 * data2['SAT']

fig = plt.plot(data2['SAT'], yhat_no, lw=2, c='#006837')
fig = plt.plot(data2['SAT'], yhat_yes, lw=2, c='#a50026')

plt.xlabel
```
